5- Calcul des statistiques descriptives

● Utilisez des fonctions de pandas pour calculer la moyenne, la médiane,
 l'écart-type, etc.


```python
import pandas as pd
```

```python
# Moyenne (Mean)
mean = df.mean()
print("Moyenne:", mean)
```

```python
# Médiane (Median)
median = df.median()
print("Médiane:", median)
```

```python
# Mode (Mode)
mode = df.mode()
print("Mode:", mode)
```

```python
# Ecart-type (Standard Deviation)
std_dev = df.std()
print("Ecart-type:", std_dev)
```

```python
# Valeur absolue moyenne (Mean Absolute Deviation)
mad = df.mad()
print("Valeur absolue moyenne:", mad)
```

```python
# Variance
variance = df.var()
print("Variance:", variance)
```

```python
# Coefficient de variance (Coefficient of Variation)
cov = df.std() / df.mean()
print("Coefficient de variance:", cov)
```

```python
# Quantiles
quantiles = df.quantile([0.25, 0.5, 0.75])
print("Quantiles:", quantiles)
```

 6-Visualisation des résultats :
 
 ● Utilisez des bibliothèques telles que Matplotlib ou Seaborn pour créer
 des graphiques représentatifs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.Series([10, 20, 30, 40, 50])
```

```python
# Création d'un histogramme
plt.hist(df, bins=5, edgecolor='black')
plt.title('Histogramme')
plt.xlabel('Valeurs')
plt.ylabel('Fréquence')
plt.show()
```

```python
# Création d'un diagramme de boxplot
sns.boxplot(df=df)
plt.title('Diagramme de boxplot')
plt.show()
```
